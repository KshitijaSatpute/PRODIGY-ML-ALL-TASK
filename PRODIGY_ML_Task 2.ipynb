```python
# Importing necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Load the dataset (update the path to your downloaded dataset)
df = pd.read_csv('C:\\Users\\Amruta\\Desktop\\PRODIGY\\Mall_Customers.csv')
```

```python
#print(df.columns)
```

```python
  # Remove leading/trailing spaces from column names
df.columns = df.columns.str.strip()
```

```python

# Selecting features for clustering with correct column names
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
# Determining the optimal number of clusters using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
# Plotting the elbow curve
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Applying K-Means with optimal clusters (choose based on the elbow curve)
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)
```

```python
# Visualizing the clusters
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y_kmeans, cmap='rainbow')
plt.title('Customer Segmentation')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
